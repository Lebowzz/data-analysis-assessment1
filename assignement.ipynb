```python
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.read_csv('./Electric_Vehicle_Population_Data.csv')
print(len(dataframe))

df_sample = dataframe.head(50) 
# i made a sample since the dataset is too large (only to make the graph clearer for informal analysis)

plt.figure(figsize=(12, 8))
df_sample['Make'].value_counts().plot(kind='bar')
plt.xlabel('Car Brand')
plt.ylabel('Count')
plt.title('Distribution of Car Brands')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 8))
df_sample['Model'].value_counts().plot(kind='bar')
plt.xlabel('Model')
plt.ylabel('Count')
plt.title('Distribution of Models')
plt.xticks(rotation=45)
plt.show()
```
